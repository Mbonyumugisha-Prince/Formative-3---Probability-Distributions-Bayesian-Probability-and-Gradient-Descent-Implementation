Gradient Descent in Codes
We are given points (1, 3) and (3, 6), m=-1, b=1, alpha=0.1

```python
from scipy.optimize import approx_fprime
import matplotlib.pyplot as plt
```

```python
x_data = [1, 3]
y_data = [3, 6]
n = len(x_data)

m = -1.0
b = 1.0
alpha = 0.1
num_iterations = 4

m_history = [m]
b_history = [b]
error_history = []
```

```python
# Mean Squared Error function
def compute_mse(params):
    m_val, b_val = params[0], params[1]
    total = 0
    for i in range(n):
        y_pred = m_val * x_data[i] + b_val
        total += (y_pred - y_data[i]) ** 2
    return total / n
```

```python
# Gradient Descent loop
print("GRADIENT DESCENT STEP BY STEP")

for iteration in range(num_iterations):
    print(f"Iteration {iteration + 1}")
    print(f"Current m = {m:.4f}, b = {b:.4f}")

    # Step 1: Compute prediction
    y_pred = [m * x_data[i] + b for i in range(n)]
    print(f"Predictions: y1 = {y_pred[0]:.4f}, y2 = {y_pred[1]:.4f}")

    # Step 2: Compute errors
    errors = [y_pred[i] - y_data[i] for i in range(n)]
    print(f"Errors: e1 = {errors[0]:.4f}, e2 = {errors[1]:.4f}")

    # Step 3: Mean Squared Error
    params = [m, b]
    mse = compute_mse(params)
    print(f"MSE = {mse:.4f}")

    # Step 4: Compute gradients
    gradients = approx_fprime(params, compute_mse, epsilon=1e-6)
    grad_m = gradients[0]
    grad_b = gradients[1]
    print(f"Gradient w.r.t m: {grad_m:.4f}")
    print(f"Gradient w.r.t b: {grad_b:.4f}")

    # Step 5: Update m and b
    m = m - alpha * grad_m
    b = b - alpha * grad_b
    print(f" Updated m = {m:.4f}, b = {b:.4f}")

    # Store history
    m_history.append(m)
    b_history.append(b)
    error_history.append(mse)

# Final Predictions using new m and b
print("FINAL RESULTS")
print(f"Final m = {m:.4f}")
print(f"Final b = {b:.4f}")

for i in range(n):
    y_final = m * x_data[i] + b
    print(f"Prediction for x = {x_data[i]}: y = {y_final:.4f} (actual y = {y_data[i]})")



```

```python
# Plot m and b over iterations

iterations_axis = list(range(num_iterations + 1))

plt.figure(figsize=(8, 5))
plt.plot(iterations_axis, m_history, marker='o', color='steelblue', linewidth=2, label='m (slope)')
plt.plot(iterations_axis, b_history, marker='s', color='darkorange', linewidth=2, label='b (intercept)')
plt.title('m and b over Iterations', fontsize=14, fontweight='bold')
plt.xlabel('Iteration')
plt.ylabel('Parameter Value')
plt.xticks(iterations_axis)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Plot error (MSE) over iterations
error_axis = list(range(1, num_iterations + 1))

plt.figure(figsize=(8, 5))
plt.plot(error_axis, error_history, marker='o', color='crimson', linewidth=2, label='MSE Error')
plt.title('Error (MSE) over Iterations', fontsize=14, fontweight='bold')
plt.xlabel('Iteration')
plt.ylabel('MSE')
plt.xticks(error_axis)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
